```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
pd.set_option('display.max_columns', 200)
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.columns.values
```

```python
df = df[['survived', 'pclass',
    'sex', 'age', #'sibsp', 'parch',
         'fare',
       #'embarked', 'class',
    #'who', 'adult_male', 'deck', 'embark_town',
       #'alive',
         'alone'
    ]].copy()
```

```python
df.head()
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df['alone'] = df['alone'].astype('int64')
```

```python
df.describe()
```

```python
sns.countplot(x='survived', data=df, hue = 'sex')
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', hue = 'pclass', data=df)
plt.show()
```

```python
sns.lineplot(x='age', y='fare', hue = 'sex' ,data=df)
plt.show()
```
